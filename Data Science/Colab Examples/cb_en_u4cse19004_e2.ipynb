```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
columns = ['Age','Workclass','fnlgwt','Education','Education num','Marital Status',
           'Occupation','Relationship','Race','Sex','Capital Gain','Capital Loss',
           'Hours/Week','Native country','Income']
train = pd.read_csv('https://raw.githubusercontent.com/vivek2319/Predicting-US-Census-Income/master/Dataset/adult-training.csv', names=columns)
```

```
train.head()
```

# Q1

```
sns.scatterplot(x='Capital Loss',y='Capital Gain',data=train,hue='Marital Status',size='Age')
```

Inference: This graph shows the capital gain and capital loss comparison with restpect to marital status and age.

# Q2

```
sns.lineplot(x="Age", y="Hours/Week",color='green',data=train,marker='<')
```

Inference: As the age increases the numeber of hours per week decreases from 40 to 80. Form 10 to 25 it increases.Its almost at the samee level for age 30 to 60.

# Q3

```
sns.boxplot(x='Age', y='Marital Status', data=train)
sns.stripplot(x='Age', y='Marital Status', data=train)
```

Inference: stripplot to show the how data is spread and boc plot to detect outliers. This graph shows the marital status with respect to age

# Q4

```
plt.figure(figsize=(10,10))
sns.stripplot(x='Relationship', y='Age',data=train,jitter=0.4)
sns.violinplot(x='Relationship', y='Age',data=train)
```

Inference:This graph shows the realtion ship and age with how spread the data is and the violin plot shows the range at that particular age. 

# Q5

```
sns.boxplot(x='Marital Status', y='Age', data=train, hue='Sex')
plt.legend(bbox_to_anchor=(1.02, 0.55), loc='upper left', borderaxespad=0)
```

Inference: This box plot between Age and Marital Status shows the outliers, never married and Married have the highest outliers.

# Q7

```
fig=sns.barplot(x=train['Income'],y=train['Age'],hue=train['Sex'],dodge=False)
plt.title('Age and Income',color='white',size=30)
plt.legend(bbox_to_anchor=[1.1, 0.5], loc='center');
plt.show()
```

Inference: Bar graph to show the income with respect to age and difference in gender.

# Q8

```
sns.catplot(x="Age", y="Marital Status", hue="Income",col="Sex", aspect=.6,kind="box", data=train);
```

Inference: This graph shows the age outliers for both male and female, based on marital status. Orange indicates income greater than 50k, blue >= 50k.
