```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(7,7))
sns.violinplot(x='Battery Type Desc',y='Batt Specific Energy (Watt-hr/kg)',data=df)
```

```python
sns.violinplot(x='Drive Desc',y='Carline Class',data=df,inner='quartile')
```

```python
sns.violinplot(x='Drive Desc',y='Carline Class',data=df,inner=None)
```

```python
sns.violinplot(y='Drive Desc',x='Carline Class',data=df,inner=None)
```

```python
sns.violinplot(x='Drive Desc',y='Carline Class',data=df,inner=None,cut=2,scale='count')
```

```python
sns.violinplot(x='Drive Desc',y='Carline Class',data=df,cut=2,bw=0.2)
```

```python
sns.violinplot(x='Drive Desc',y='Carline Class',data=df,inner=None,hue='Trans')

```

```python
sns.violinplot(y='Drive Desc',x='Carline Class',data=df,linewidth=4)
```

```python
sns.violinplot(y='Drive Desc',x='Carline Class',data=df,palette='pastel')
```

```python
sns.violinplot(y='Drive Desc',x='Carline Class',data=df,saturation=0.1)
```

```python
sns.violinplot(y='Drive Desc',x='Carline Class',data=df,color='red')
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(x='Drive Desc',y='Displacement',data=df)
sns.violinplot(x='Drive Desc',y='Displacement',data=df)
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(x='Drive Desc',y='Displacement',data=df,palette='Set1')
sns.violinplot(x='Drive Desc',y='Displacement',data=df,palette='pastel')
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(x='Drive Desc',y='Displacement',data=df,linewidth=1)
sns.violinplot(x='Drive Desc',y='Displacement',data=df)
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(x='Drive Desc',y='Displacement',data=df,linewidth=1)
sns.violinplot(x='Drive Desc',y='Displacement',data=df,inner = None)
```

```python
plt.figure(figsize=(50,20))
sns.stripplot(x='Carline Class Desc',y='Comb FE (Guide) - Conventional Fuel',data=df,linewidth=2)
sns.violinplot(x='Carline Class Desc',y='Comb FE (Guide) - Conventional Fuel',data=df)
```

```python
plt.figure(figsize=(50,20))
sns.stripplot(x='Carline Class Desc',y='Comb FE (Guide) - Conventional Fuel',data=df,linewidth=2)
sns.violinplot(x='Carline Class Desc',y='Comb FE (Guide) - Conventional Fuel',data=df,hue='Trans')
```

```python
plt.figure(figsize=(50,20))
sns.stripplot(x='Carline Class Desc',y='Comb FE (Guide) - Conventional Fuel',data=df,linewidth=1,palette='Set1')
sns.violinplot(x='Carline Class Desc',y='Comb FE (Guide) - Conventional Fuel',data=df,palette='pastel',inner=None)
```
